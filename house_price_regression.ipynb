```python
X = housing.drop("median_house_value",axis = 1)
y = housing["median_house_value"].copy()
```

```python
from sklearn.model_selection import train_test_split
```

```python
np.random.seed(42)
X_train,X_test,y_train,y_test =train_test_split(X,y,stratify= housing["income_cat"],test_size = 0.2)
```

```python
X_train = X_train.drop("income_cat",axis = 1)
X_test= X_test.drop("income_cat", axis = 1)
```

```python
cat_features = X_train.select_dtypes(exclude=[np.number]).columns

num_features = X_train.select_dtypes(include =[np.number]).columns
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer
```

```python
from sklearn.preprocessing import MinMaxScaler, StandardScaler
min_max = MinMaxScaler()

cat_pipeline = Pipeline([
    ("imputer",SimpleImputer(strategy="most_frequent")),
    ("onehot", OneHotEncoder())
])
num_pipeline = Pipeline([
    ("imputer",SimpleImputer(strategy = "median")),
    ('standart_scaler',MinMaxScaler())
])
transformer = ColumnTransformer([
    ("cat_transformer",cat_pipeline,cat_features),
    ("num_transformer",num_pipeline,num_features)
],remainder = "passthrough")

```

```python
X_train_transform = transformer.fit_transform(X_train)
X_train_transform_df = pd.DataFrame(X_train_transform)
```

```python
X_test_transform = transformer.transform(X_test)
X_test_transform_df = pd.DataFrame(X_test_transform)
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rnd_reg = RandomForestRegressor()
```

```python
rnd_reg.fit(X_train_transform_df,y_train)
```

```python
rnd_reg.score(X_test_transform_df,y_test)
```

```python
np.random.seed(42)
full_pipeline = Pipeline([
    ("preprocessing",transformer),
    ("random_forest",rnd_reg)
])
full_pipeline.fit(X_train,y_train)
```

```python
full_pipeline.score(X_test,y_test)
```

```python
from sklearn.model_selection import cross_val_score
cross_val_score(full_pipeline,X,y,cv=5)
```
