```python
import numpy as np

np.random.seed(42)
m = 100
X = 2*np.random.rand(m,1)
y = 4+3*X + np.random.randn(m,1)
```

```python
plt.figure(figsize = (6,4))
plt.plot(X,y,"b.")
plt.xlabel("$x_1$")
plt.ylabel("$y$",rotation = 0)
plt.axis([0,2,0,15])
plt.grid()
plt.show()
```

```python
matrix_a = np.array([[1,2],[3,4]])
matrix_b = ([[5,6],[7,8]])

result1 = np.dot(matrix_a,matrix_b)
result1
```

```python
from sklearn.preprocessing import add_dummy_feature
X_b = add_dummy_feature(X)

theta_best = np.linalg.inv(X_b.T @ X_b)@ X_b.T @y
```

```python
theta_best
```

```python
X_new = np.array([[0],[2]])
X_new_b = add_dummy_feature(X_new)
y_predict = X_new_b @ theta_best
y_predict
```

```python
plt.figure(figsize = (6,4))
plt.plot(X_new,y_predict,"r-",label = "predictions")
plt.plot(X,y,"b.")
plt.xlabel("$x_1$")
plt.ylabel("$y$",rotation = 0)
plt.axis([0,2,0,15])
plt.grid()

plt.legend(loc = "upper left")
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
lnr = LinearRegression()
lnr.fit(X,y)
lnr.intercept_,lnr.coef_
```

#Stochastic Gradient Dascent

```python
np.random.seed(42)
eta = 0.1
m = len(X_b)
theta = np.random.randn(2,1)

for epoch in range (n_epochs):
  for iteration in range(m):
    random_index = np.random.randint(m)
    xi=X_b[random_index:random_index+1]
    yi = y[random_index:random_index+1]
    gradients = 2*xi.T@(xi@theta-yi)
    theta = theta-eta*gradients
```

#Polynomial Regressiom

```python
np.random.seed(42)
m=100
X=6*np.random.rand(m,1)-3
y= 0.5* X ** 2+X+2+np.random.randn(m,1)
```

```python
plt.figure(figsize=(6,4))
plt.plot(X,y,"b.")
plt.xlabel("$x_1$")
plt.ylabel("$y$",rotation = 0)
plt.axis([-3,3,0,10])
plt.grid()
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
lnr_reg = LinearRegression()
lnr_reg.fit(X,y)
```

```python
X_new = [[-3],[3]]
X_preds = lnr_reg.predict(X_new)
X_preds
```

```python
plt.figure(figsize=(6,4))
plt.plot(X,y,"b.")
plt.plot(X_new,X_preds,"r-")
plt.xlabel("$x_1$")
plt.ylabel("$y$",rotation = 0)
plt.axis([-3,3,0,10])
plt.grid()
plt.show()
```

```python
from sklearn.preprocessing import PolynomialFeatures

poly_features =PolynomialFeatures(degree = 2,include_bias = False)
X_poly = poly_features.fit_transform(X)
X[0]
```

```python
X_poly[0]
```

```python
lnr_reg = LinearRegression()
lnr_reg.fit(X_poly,y)
lnr_reg.intercept_,lnr_reg.coef_
```

```python
X_poly.shape
```

```python
X_new = np.linspace(-3,3,100).reshape(100,1)
X_new_poly = poly_features.transform(X_new)
y_new = lnr_reg.predict(X_new_poly)

plt.figure(figsize = (6,4))
plt.plot(X,y,"b.")
plt.plot(X_new,y_new,"r-",linewidth =2,label = "Predictions")
plt.xlabel("$x_1$")
plt.ylabel("$y$",rotation = 0)
plt.legend(loc ="upper left")
plt.axis([-3,3,0,10])
plt.grid()
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline

plt.figure(figsize = (6,4))
for style,width,degree in (("r-+",2,1),("b--",2,2),("g-",1,300)):
  polybig_features=PolynomialFeatures(degree=degree,include_bias=False)
  std_scaler = StandardScaler()
  lnr_reg =LinearRegression()
  polynomial_regression =make_pipeline(polybig_features ,std_scaler ,lnr_reg)
  polynomial_regression.fit(X,y)
  y_newbig = polynomial_regression.predict(X_new)
  label = f"{degree} degree{'s'if degree > 1 else ''}"
  plt.plot(X_new,y_newbig,style,label = label, linewidth =width)

plt.plot(X,y, "b.",linewidth = 3)
plt.legend(loc ="upper left")
plt.xlabel("$x_1$")
plt.ylabel("$y$",rotation = 0)
plt.axis([-3,3,0,10])
plt.grid()
plt.show()
```

```python
from sklearn.model_selection import learning_curve

train_sizes,train_scores,valid_scores =learning_curve(
    LinearRegression(),X,y,train_sizes = np.linspace(0.001,1.0,40),cv = 5,
    scoring = "neg_root_mean_squared_error"
)
train_errors = -train_scores.mean(axis = 1)
valid_errors = -valid_scores.mean(axis = 1)

plt.figure(figsize = (6,4))
plt.plot(train_sizes,train_errors,"r-+",linewidth = 2,label = "train")
plt.plot(train_sizes,valid_errors,"b-",linewidth = 3,label = "valid")

plt.xlabel("training set size")
plt.ylabel("RMSE")

plt.grid()
plt.legend(loc = "upper right")
plt.axis([0,80,0,2.5])

plt.show()
```

```python
from sklearn.pipeline import make_pipeline

polinomial_regression = make_pipeline(
    PolynomialFeatures(degree=2, include_bias =False),
    LinearRegression()
)
train_sizes,train_scores,valid_scores =learning_curve(
    polinomial_regression,X,y,train_sizes = np.linspace(0.001,1.0,40),cv = 5,
    scoring = "neg_root_mean_squared_error"
)
```

```python
train_errors = -train_scores.mean(axis = 1)
valid_errors = -valid_scores.mean(axis = 1)

plt.figure(figsize = (6,4))
plt.plot(train_sizes,train_errors,"r-+",linewidth = 2,label = "train")
plt.plot(train_sizes,valid_errors,"b-",linewidth = 3,label = "valid")

plt.xlabel("training set size")
plt.ylabel("RMSE")

plt.grid()
plt.legend(loc = "upper right")
plt.axis([0,80,0,2.5])

plt.show()
```

```python
plt.figure(figsize = (6,4))
plt.plot(X,y, ".")
plt.xlabel("$X_1$")
plt.ylabel("y",rotation = 0)
plt.axis([0,3,0,3.5])
plt.grid()
plt.show()
```

```python
from sklearn.linear_model import Ridge

ridge_reg = Ridge(alpha = 0.1,solver = "cholesky")
ridge_reg.fit(X,y)
ridge_reg.predict([[1.5]])
```

```python
sgd_reg = SGDRegressor(penalty = "l2",
                       max_iter = 1000,eta0 = 0.01,random_state = 42)
sgd_reg.fit(X,y.ravel())
sgd_reg.predict([[1.5]])
```

```python
def plot_model(model_class, polynomial, alphas, **model_kwargs):
    plt.plot(X, y, "b.", linewidth=3)
    for alpha, style in zip(alphas, ("b:", "g--", "r-")):
        if alpha > 0:
            model = model_class(alpha, **model_kwargs)
        else:
            model = LinearRegression()
        if polynomial:
            model = make_pipeline(
                PolynomialFeatures(degree=10, include_bias=False),
                StandardScaler(),
                model)
        model.fit(X, y)
        y_new_regul = model.predict(X_new)
        plt.plot(X_new, y_new_regul, style, linewidth=2,
                 label=fr"$\alpha = {alpha}$")
    plt.legend(loc="upper left")
    plt.xlabel("$x_1$")
    plt.axis([0, 3, 0, 3.5])
    plt.grid()
plt.figure(figsize=(9, 3.5))
plt.subplot(121)
plot_model(Ridge, polynomial=False, alphas=(0, 10, 100), random_state=42)
plt.ylabel("$y$  ", rotation=0)
plt.subplot(122)
plot_model(Ridge, polynomial=True, alphas=(0, 10**-5, 1), random_state=42)
plt.show()
```

```python
plt.figure(figsize = (9,3.5))
plt.subplot(121)
plot_model(Lasso, polynomial = False, alphas = (0,0.1,1),random_state = 42)
plt.ylabel("$y$",rotation = 0)
plt.subplot(122)
plot_model(Lasso,polynomial = True,alphas = (0,1e-2,1),random_state = 42)
plt.show()
```
